# Summary 
Out of a total of 891 passengers:
- 577 were men. Of those 577 men only 109 survived
- 341 were women. Of those 341 women, 233 survived.
- Of the 577 men, 58 were boys and 23 of them survived
- Of the 341 women, 55 were girls and 38 survived

```python
import matplotlib.pyplot as plt
plt.pie([519,58,286,55],labels=["Man","Boys", "Woman", "Girl"],autopct="%0.1f %%")
plt.title("Passengers sex distribution")
plt.show()
```
