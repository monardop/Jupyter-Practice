### En cuanto a horarios 
El primer cuatrimestre del año 2019 es, con bastante diferencia, el período más peligroso. De 17  a 20 hs es el horario donde suceden la gran mayoría de crímenes, esto coincide con el momento del día donde más tráfico tiende a haber, tanto de peatones como de automóviles. A su vez, el día con mayor registro de denuncias es el viernes. Los fines de semana parecen ser mucho más tranquilos a diferencia de la semana laboral general (Lunes a viernes)
### En cuanto a demografía
La amplia mayoría de crímenes fueron realizados en Palermo, Balvanera, San Nicolás, Recoleta y Flores; lugares donde suele haber mucho trafico y turismo. Por el contrario, los barrios con menos reportes son barrios con densidad poblacional muy baja y, a su vez, con barrios privados, lo que lóguicamente dificultaría mucho los hechos delictivos menores como hurtos y robos.
### En cuanto a tipología
El 95.3% de los crímenes registrados son robos. El 4.7% de estos crímenes pasaron a mayores, siendo un 0.2% homicidios y el resto incluyen lesiones a la víctima.
